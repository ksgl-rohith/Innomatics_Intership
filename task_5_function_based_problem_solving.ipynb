```python
#Problem 1: Social Media – Post Engagement Analyzer
def engagement_score(likes):
  #calculate total likes using loop
  total = 0
  for i in likes:
    total += i
  #conditional statement for requirement
  if total >= 1000:
    print("Post Status: Viral Post")
  else:
    print("Post Status: Normal Engagement")
  return
#lets consider the likes in the form of list
likes = [100,20,300,400,578,30,25,44,612,70,88]
engagement_score(likes)
```

```python
#Problem 2: Healthcare – Medicine Stock Alert System
def med_stock_levels(stocks):
  #consider the stocks in the dictionary format
  for med,stock in stocks.items():
    #using conditional statement print stock and medicine
    if stock < 10:
      print(f"Medicine Stock: {stock}")
      print(f"Status: Low Stock! Need more {med} medicine")
    else:
      print(f"Medicine Stock: {stock}")
      print(f"Status: Stock {med} is sufficient")
#stocks and medicines

stocks = {
    "Dolo" : 9,
    "Paracetemol" : 11,
    "Zolo" : 15,
    "Cheston cold" : 8
}

med_stock_levels(stocks)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def seasonal_rainfall(rainfall,required_level):
  #calculating the total rainfall
  total = 0
  for i in rainfall:
    total += i
  #average rainfall
  avg_rainfall = total/len(rainfall)
  #conditional logic
  if avg_rainfall >= required_level:
    print(f"Average Rainfall: {avg_rainfall}")
    print("Rainfall Status: Adequate Rainfall")
  else:
     print(f"Average Rainfall: {avg_rainfall}")
     print("Rainfall Status: inadequate Rainfall")

rainfall = [120,100,10,22,31,80,110] #assume this data
required_level = 60 #assume required level to be 60
seasonal_rainfall(rainfall,required_level)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
def duplicate(usernames):
  #create a set to store the unique usernames
  my_set = set()
  #check whether they are present or not
  for username in usernames:
    if username in my_set:
      print(f"Duplicate Accounts Found for {username}: Yes")
    else:
      #add if its not present
      print(f"Duplicate Accounts Found for {username}: No")
      my_set.add(username)

usernames = ["sunny","lucky","sunny","rocky","sweety"]
duplicate(usernames)
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker
def check_eligibility(age):
  #conditional logic
  if age>= 18:
    print(f"Patient Age: {age}")
    print("Eligiblity Status: Eligible")
  else:
    print(f"Patient Age: {age}")
    print("Eligiblity Status: Not Eligible")

age = int(input())
check_eligibility(age)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
def filter_crop_price(prices):
  premium_crop = [] #list to store all the premium prices
  for price in prices:
    #if crop is premium then add it to the list
    if price > 2000:
      premium_crop.append(price)
  #print the list to get the premium crops
  print(f"Premium Crops: {premium_crop}")

prices = [1200,2500,1100,3000,4000,2800]
filter_crop_price(prices)
```

```python
#Problem 7: System Monitoring – Application Health Checker
def system_monitoring(err_count):
  #conditional logic
  if err_count == 0:
    print(f"Error Count: {err_count}")
    print("System Status: Healthy")
  elif err_count <= 5:
    print(f"Error Count: {err_count}")
    print("System Status: Minor Issues")
  elif err_count > 5:
    print(f"Error Count: {err_count}")
    print("System Status: Critical Issues")

err_count = int(input())
system_monitoring(err_count)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker
def daily_limit_checker(amount):
  daily_limit = 50000 #assume the daily limit as 50000
  #conditional logic
  if amount <= daily_limit :
    print(f"Transaction Amount: {amount}")
    print("Transaction Status: Approved")
  elif amount > daily_limit :
    print(f"Transaction Amount: {amount}")
    print("Transaction Status: Rejected")

amount = int(input())
daily_limit_checker(amount)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_eligibility(attendance):
  #conditional logic
  if attendance >= 75:
    print(f"Attendance Percentage: {attendance}")
    print("Exam Eligibity: Eligible")
  else:
    print(f"Attendance Percentage: {attendance}")
    print("Exam Eligibity: Not Eligible")

attendance = int(input())
attendance_eligibility(attendance)
```

```python
#Problem 10: Smart Electricity Bill Analyzer
def electricity_bill(units):
  remaining = units #for track the units
  bill = 0 #for calculating the bill

  #slab 1 between 0 to 100 or more
  if remaining > 100:
    bill = 100 * 3
    remaining -= 100
  else:
    bill += remaining * 3
    remaining = 0

  #slab 2 between 101 to 200 or more
  if remaining > 100:
    bill += 100 * 5
    remaining -= 100
  else:
    bill += remaining * 5
    remaining = remaining - 100

  #slab 3 more than 200
  if remaining > 0:
    bill += remaining * 7

  #usage classification
  if bill < 500:
    usage = "Low Usage"
  elif bill >= 500 and bill < 1500:
    usage = "Moderate Usage"
  else:
    usage = "High Usage"

  return bill,usage

units = int(input())
print(electricity_bill(units))
```
