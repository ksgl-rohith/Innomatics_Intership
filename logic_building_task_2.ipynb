```python
#Problem 1: Unique Words in a Sentence
sentence = input("Enter sentence: ")
#Converting the list into set to remove the duplicates
words = set(sentence.split(" "))
count = 0
#Counting the words in set
for word in words:
  count+=1
#Displaying the output
print("Unique words count: ", count)
print("Unique words: ", words)
```

```python
#Problem 2: Highest Salary from Employee Data

#problem contains the input in dict so converting the list to dictionary
employee_names = input()
employee_salaries = input()
#converting into list
employee_name = list(employee_names.split(" "))
employee_salary = list(map(int,employee_salaries.split(" ")))
#initialize the dictionary
employee_dict = {}
max_salary = employee_salary[0]
key_to_print = employee_name[0]

for employee in range(len(employee_name)):
  #adding key and value pairs to dictionary
  employee_dict[employee_name[employee]] = employee_salary[employee]
  #calculating the highest salary
  if employee_salary[employee] > max_salary:
    max = employee_salary[employee]
    key_to_print = employee_name[employee]
#displaying the output
print("Highest Salary: ",key_to_print,"-", employee_dict[key_to_print] )
```

```python
#Problem 3: Find Maximum and Minimum Values

numbers = input().split()
numbers = list(map(int, numbers))
#initialise with first value
max_number = numbers[0]
min_number = numbers[0]
#looping through the numbers
for number in numbers:
  #calculating the max value
  if max_number<number:
    max_number = number
  #calculating the min value
  if min_number>number:
    min_number = number
#displaying the output
print("List:",numbers)
print(f"Max: {max_number}")
print(f"Min: {min_number}")
```

```python
#Problem 4: Count Products Above a Price Threshold
prices = input().split()
prices = list(map(int,prices))
#looping through the prices
count = 0
for price in prices:
  #calculating the max price
  if price>1000:
    count+=1
#displaying the output
print("Products above 1000:", count)
```

```python
#Problem 5: Calculate Attendance Percentage
attendance = input().split()
#looping through the attendance
count = 0
for value in attendance:
  #if value equals present then counting the values
  if value == "P":
    count+=1
total_days = len(attendance)
#Formula for total attendance
attendance_percentage = (count/total_days)*100
#displaying the output
print("Attendance Percentage:", attendance_percentage)
```

```python
#Problem 6: Remove Duplicate Phone Numbers
phone_numbers = input().split()
#converting the string to int and list to set
phone_numbers = set(map(int,phone_numbers))
#displaying the output
print("Unique phone numbers:",phone_numbers)
```

```python
#Problem 7: Count Character Frequency
string = input()
char_dict = {}
#looping through character in given string
for char in string:
  #condition if char is in dictionary or not
  if char in char_dict:
    char_dict[char] += 1
  else:
    char_dict[char] = 1
#displaying the output
print(char_dict)
```

```python
#Problem 8: Convert List to Tuple
numbers = input().split()
#converting the list to tuple by using the tuple()
numbers = tuple(map(int,numbers))
#displaying the output
print(numbers)
```

```python
#Problem 9: Check if a Key Exists in a Dictionary
employees = input().split()
employee_id = input().split()
#looping for adding the key value pair in the dictionary
emp_dict = {}
for i in range(len(employees)):
  emp_dict[employee_id[i]] = employees[i]
#read a key value
check_emp = input("Enter the employee id: ")
#check whether it is present or not
if check_emp in emp_dict:
  print("Employee exists")
else:
  print("Employee doesn't exist")

```

```python
#Problem 10: Calculate Average Marks
marks = input().split()
marks = list(map(int,marks))
#calculating the sum using sum() function
sum_marks = sum(marks)
#assume total marks to be 600
total_marks = 600
#rounding the decimal values and calucaling the average
avg_marks = round((sum_marks/total_marks)*100,2)
#displaying the output
print("Average Marks:", avg_marks)
```
