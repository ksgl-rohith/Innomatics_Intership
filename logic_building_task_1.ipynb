```python
#1. User login check
username = "admin" #given username
password = "1234" #given password

#logic for the login
if username == "admin" and password == "1234":
  print("Login Successful")
else:
  print("Invalid Credentials")
```

```python
#2. Pass/Fail analyzer
marks = [45,78,90,33,60] #given data

#logic for the conditions
pass_count=0
fail_count=0
for mark in marks:
  if mark>=50:
    pass_count+=1
  else:
    fail_count+=1

#printing the output
print("Total pass students count: ",pass_count)
print("Total fail students count: ",fail_count)
```

```python
#3. Simple data cleaner
names = [" Alice ", "bob", " CHARLIE "]

#Logic for cleanig the data
cleaned_names = [name.strip().lower() for name in names]

#printing the final output
print(cleaned_names)
```

```python
#4. Message Length Analyzer
messages = ["Hi", "Welcome to the platform", "OK"]

#logic for the length and flag check
for message in messages:
  if len(message)>10:
    flag = True
  else:
    flag = False

  #printing the final output
  print("The message : '",message,"' length is ",len(message)," flag is ", flag)
```

```python
#5. Error Message Detector
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count=0
#Logic
for log in logs:
  if log == "ERROR":
    count+=1

#printing the final output
print(count)
```
