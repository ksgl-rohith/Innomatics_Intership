```python
#Problem 1: Mobile Recharge Validation System
def mobile_recharge(recharge_amt,plan):
  if recharge_amt<50:
    print("Recharge amount must be greater ₹50")
    return
  status = False
  while status==False:
    if plan in [199,299,399,599] and recharge_amt - plan >= 0:
      print("Recharge request is successful")
      print(f"Remaining balance: {recharge_amt - plan}")
      status=True
    elif plan in [199,299,399,599] and recharge_amt - plan < 0 and plan!=199:
      print("Recharge amount is not sufficient for the plan")
      user = input("Would you like to change the plan?[Yes or No]: ")
      if user == "Yes" or user == "yes":
        plan = int(input("Enter a valid plan: "))
        status = False
      elif user == "No" or user == "no":
        return "Thank You!"
    elif plan in [199,299,399,599] and recharge_amt - plan < 0 and plan==199:
      print("Insufficient funds to recharge")
      status = True
    elif plan not in [199,299,399,599]:
      plan = int(input("Enter a valid plan from [199,299,399,599]: "))
      status = False
  return "Thank You!"

recharge_amt = int(input("Enter the recharge amount: "))
plan = int(input("Enter the plan from [199,299,399,599]: "))
mobile_recharge(recharge_amt,plan)
```

```python
#Problem 2: Inventory Reorder Alert System
def inventory_alert_system(my_dict):
  for product,stock in my_dict.items():
    if stock < 15:
      print(f"Reorder the {product} of stock {stock}.")
    else:
      print(f"{product} Stock OK")
  return

my_dict = {
    "Rice" : 10,
    "Maida": 25,
    "Whole Wheat" : 13,
    "Ragi flour" : 15
}

inventory_alert_system(my_dict)
```

```python
#Problem 3: Student Result Processing System
def student_result(marks):
  #average marks
  avg_marks = sum(marks)/len(marks)
  #result
  if avg_marks >= 50:
    res = "Pass"
  else:
    res = "Fail"
  #return result
  return res,avg_marks

marks = list(map(int,input().split()))
print(student_result(marks))
```

```python
#Problem 4: Cab Fare Estimator with Retry Option
def cab_fare_estimator(km,peak_hour):
  status = False
  while status == False:
    #check if it satisfies the base fare
    if peak_hour:
      if  (km * 12) * 1.25 < 50:
        print("The base fare is 50. So the fare is 50. Is it ok?[Yes or No]")
        base = input()
        if base == "Yes":
          fare = 50
          status = True
        else:
          print("Thank you for your time")
          return
        status = False
      else:
        fare = (km * 12) * 1.25
        return fare
    else:
      if  (km * 12) < 50:
        print("The base fare is 50. So the fare is 50. Is it ok?[Yes or No]")
        base = input()
        if base == "Yes":
          fare = 50
          status = True
        else:
          print("Thank you for your time")
          return
        status = False
      else:
        fare = (km * 12)
        return fare
  return

km = int(input())
peak_hour = bool(input())
print(cab_fare_estimator(km,peak_hour))




```

```python
#Problem 5: Employee Attendance Eligibility Checker
def attendance_eligibility_checker(attendance):
  count = 0
  for status in attendance:
    if status == "P":
      count += 1
  if count/len(attendance) > 0.75:
    res = "Eligible"
  else:
    res = "Not Eligible"
  return res

attendance = input().split()
print(attendance_eligibility_checker(attendance))
```

```python
#Problem 6: Password Strength Checker
def password_check(password):
  digit = False
  is_special = False
  length = False
  special = "@#$"
  if len(password) < 8:
    print("Password must be more than 8 characters.")
    length = True

  for char in password:
    if char.isdigit():
      digit = True
    if char in special:
      is_special = True

  if length and is_special and special :
    print("successful login")
  elif not digit and not is_special:
    print("Must contain a digit and special character")
  elif not digit and is_special:
    print("Must contain a digit")
  elif digit and not is_special:
    print("Must contain a special character")

  return

password = input()
password_check(password)


```
