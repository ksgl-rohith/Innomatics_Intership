```python
#Problem Statement 1: Employee Performance Bonus Eligibility
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
#create an empty list to store the multpile names
emp= []
max_score = max(employees.values())
#check if there are multiple names for the score by adding the names in the list
for name,score in employees.items():
  if score==max_score:
    emp.append(name)
#join the strings in the list
name_string = ", ".join(emp)
#displaying the output according to the format
print(f"Top Performers Eligible for Bonus: {name_string} (Score: {max_score})")

```

```python
#Problem Statement 2: Search Query Keyword Analysis
query = input()
#lower the string
query = query.lower()
import string
#punctions from the string library
for p in string.punctuation:
    query = query.replace(p, "")
#get the clean words without punctions
words = query.split()
#count the frequency
freq = {}
for word in words:
    freq[word] = freq.get(word, 0) + 1
#get the frequent words which repeat more then once
result = {}
for word, count in freq.items():
    if count > 1:
        result[word] = count
#displaying the output according to the format
print(result)

```

```python
#Problem Statement 3: Sensor Data Validation
sensor_readings = input()
sensor_readings = list(map(int,sensor_readings.split(", ")))
#Identify readings that are even numbers (valid readings) and storing it in result
res = []
for idx,value in enumerate(sensor_readings):
  if value%2==0:
    res.append((idx,value))
#displaying the output according to the format
print("Valid Sensor Readings (Hour, Value):")
print(res)
```

```python
#Problem Statement 4: Email Domain Usage Analysis
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
#create the empty domains dictionary
domains = {}
for mail in emails:
  #split it for only domains
  domain = mail.split("@")[1]
  #if domain exists then add 1 or add the domain
  if domain in domains:
    domains[domain] += 1
  else:
    domains[domain] = 1
#calculate the length of the emails
total_mails = len(emails)
for val,count in domains.items():
  #percentage logic
  percentage = (count/total_mails)*100
  #displaying the output according to the format
  print(f"{val}: {int(percentage)}%")
```

```python
#Problem Statement 5: Sales Spike Detection
sales = [1200, 1500, 900, 2200, 1400, 3000]
#avg formula
avg_sales = sum(sales)/len(sales)
for day in range(1, len(sales)):
    #if the sales are more than 30% above average then print
    if sales[day] > sales[day-1] * 1.3:
      #displaying the output according to the format
      print(f"Day {day+1}: {sales[day]}")
```

```python
#Problem Statement 6: Duplicate User ID Detection
#Counter is used to calculate the frequencies in the python
from collections import Counter
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
#use the counter and extract the count
counts = Counter(user_ids)
#loop the counts to display the values
for user, count in counts.items():
    if count > 1:
      #displaying the output according to the format
      print(f"{user} → {count} times")
```
