**Problem Statement 1:Smart Parking Lot Management System**

```python
def smart_parking(cap,logs):
  count_IN = 0
  count_OUT = 0
  for i in logs:
    if i == "IN":
      count_IN += 1
    else:
      count_OUT += 1
  total_parked = count_IN - count_OUT

  if total_parked >= cap:
    # print("Slots have been exceeded")
    status = "Not Available"
  else:
    status = "Available"
  #displays the output
  print(f"Currently Parked Vehicles: {total_parked}")
  print(f"Parking Status: {status}")

smart_parking(3,["IN","OUT","IN","IN","IN"])
print()
smart_parking(50,["IN","OUT","IN","IN","IN"])

```

**Problem Statement 2:Online Food Delivery Time Estimator**


```python
def delivery_time(dist,weather,traffic):
  #Base time to deliver in normal condition per km be 5 mins
  base = 5

  if weather == "Normal" and traffic == "Low":
    time_req = base*dist
  if weather == "Normal" and traffic == "high":
    time_req = base*dist*1.25 #if traffic is high it will delay
  elif weather == "Rainy" and traffic == "Low":
    time_req = base*dist*1.10 #if its rainy and traffic low it will take 10% more time
  elif weather == "Rainy" and traffic == "High":
    time_req = base*dist*1.50 #if its rainy and traffic high it will take 50% more time

  #displays the output
  print(f"Estimated Delivery Time: {time_req} minutes")

delivery_time(8,"Rainy","Low")
print()
delivery_time(8,"Rainy","High")
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**


```python
def booking_data(total_seats,booked_seats):
  sum_of_booked = sum(booked_seats)

  Percentage = (sum_of_booked/total_seats)*100
  if sum_of_booked == total_seats:
    status = "Housefull"
    shows = "Required"
  elif Percentage > 75:
    status = "Almost Full"
    shows = "Required"
  elif Percentage > 30:
    status = "Decent"
    shows = "Required"
  else:
    status = "Low"
    shows = "Not Required"
  #displays the output
  print(f"Occupancy: {Percentage:.2f}%")
  print(f"Show Status: {status}")
  print(f"Additional shows: {shows}")


booking_data(100,[10,1,10,1,10,1,1,1,1,1])
print()
booking_data(100,[1,1,1,1,10,1,1,1,1,1])

```


**Problem Statement 4:Cloud Server Load Classification System**

```python
def server_load(cpu_readings):
  avg_readings = (sum(cpu_readings)/(len(cpu_readings)*100))*100
  if avg_readings > 80:
    status = "Critical"
  elif avg_readings >= 50:
    status = "Warning"
  elif avg_readings < 50:
    status = "Normal"
  #displays the output
  print(f"Average CPU Load: {avg_readings}%")
  print(f"Server Status: {status}")


server_load([45, 60, 70, 85, 90])
print()
server_load([70, 90, 70, 95, 90])
```

**Problem Statement 6: Online Event Registration Capacity Controller**

```python
def event_reg( cap, reg ):
  if cap < reg:
    waitlisted_user = reg - cap
    status = "Closed"
    reg = cap
  else:
    waitlisted_user = 0
    status = "Open"
  #displays the output
  print(f"Confirmed Registrations: {reg}")
  print(f"Waitlisted Users: {waitlisted_user}")
  print(f"Registration Status: {status}")

event_reg(100,105)
print()
event_reg(100,80)
```
