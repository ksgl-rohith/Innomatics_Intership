**Problem 1: Smart Light Controller**


```python
class smart_light_control:
  #constructor
  def __init__(self,light_name,action):
    self.light_name = light_name
    self.action = action
  #status method
  def status(self):
    print(f"{self.light_name} is {self.action}")
#creating an object of smart_light_control
control = smart_light_control("Bedroom Light","ON")
control.status()
```

**Problem 2: Employee ID Card System**

```python
class emp_id:
  #constructor
  def __init__(self,name,ID,dept):
    self.name = name
    self.ID = ID
    self.dept = dept
  #details method
  def details(self):
    print("--Employee ID Card--")
    print(f"Name: {self.name}")
    print(f"ID: {self.ID}")
    print(f"Department: {self.dept}")
#creating an object of empployee ID card
emp = emp_id("Rohith","EMP 111","IT")
emp.details()
```

**Problem 3: Mobile Contact Record**

```python
class contact_record:
  #constructor
  def __init__(self,name,phn_no):
    self.name = name
    self.phn_no = phn_no
  #info method
  def info(self):
    print("--Contact Saved--")
    print(f"Contact Name: {self.name}")
    print(f"Phone Number: {self.phn_no}")
#creating an object of contact record
contact = contact_record("Anita","9988776655")
contact.info()
```

**Problem 4: Product Price Tag Generator**


```python
class price_tag_generator:
  #constructor
  def __init__(self,p_name,price):
    self.p_name = p_name
    self.price = price
  #tag method
  def tag(self):
    print("--Product Details--")
    print(f"Product: {self.p_name}")
    print(f"Price: {self.price}")
#creating an object of price_tag_generator
ptr = price_tag_generator("Headphones",2499)
ptr.tag()
```

**Problem 5: Movie Rating Display System**

```python
class movie_rating_system:
  #constructor
  def __init__(self,name,rating):
    self.name = name
    self.rating = rating
  #details method
  def details(self):
    print("--Movie Details--")
    print(f"Movie: {self.name}")
    print(f"Rating: {self.rating} / 5")
#creating an object of movie rating system
movie = movie_rating_system("Bahubali",4.9)
movie.details()
```

**Problem 6: Delivery Address Manager**

```python
class delivery_address_manager:
  #constructor
  def __init__(self,name,address):
    self.name = name
    self.address = address
  #details method
  def details(self):
    print("--Delivery Details--")
    print(f"Customer: {self.name}")
    print(f"Address: {self.address}")
#creating an object of delivery address manager
delivery = delivery_address_manager("Karthik","Delhi")
delivery.details()
```
